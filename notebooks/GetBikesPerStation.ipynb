```python

def demanda_negativa_plot(day, station):
    
    grouped_days = data.groupby(data['Inicio_del_viaje'].dt.day)

    day_n = grouped_days.get_group(day)
    
    #
    day_n_station_n = day_n.loc[day_n['Origen_Id'] == station]
    
    hourly_counts = day_n_station_n.groupby(day_n_station_n['Inicio_del_viaje'].dt.hour)['Inicio_del_viaje'].count()
    
    try:
        midnight = pd.Series([hourly_counts[0]], index = [24])
                
    except:
        
        midnight = pd.Series([0], index = [24])
        
    hours_range = pd.Series(range(5, 25))

    hourly_counts = hourly_counts.reindex(hours_range, fill_value = 0)
    
    hourly_counts = pd.concat([hourly_counts,midnight])
    
    hourly_counts_df = pd.DataFrame({'Hour': hourly_counts.index, 'Count': hourly_counts.values})

    # Plot
    sns.set(style="whitegrid")
    plt.figure(figsize=(12, 6))
    ax = sns.lineplot(x='Hour', y='Count', data=hourly_counts_df, marker='o')
    #plt.title(f'Demanda Negativa (d-) el día {day} de Enero de 2023 en la estación {station}')
    
    plt.title(
    f'Demanda Negativa (d-) el día {day} de Enero de 2023 en la estación {station}',
    color='black',
    fontsize=12,
    #loc='center',
    fontweight='bold',
    style='normal',
    family='Arial',
    y=1.00)

    plt.xlabel('Hora')
    plt.ylabel('Demanda (d-)')

    ax.set_yticks(list(range(0, max(hourly_counts.values)+1)))
    ax.set_xticks(hourly_counts.index)
    

    plt.axvline(x=5, color='red', linestyle='--', alpha = 0.3, label='Horarios de Rebalanceo')
    plt.axvline(x=12, color='red', linestyle='--', alpha = 0.3)
    plt.axvline(x=19, color='red', linestyle='--', alpha = 0.3)
    plt.axvline(x=24, color='red', linestyle='--', alpha = 0.3)
    plt.legend(loc='upper left')
    #plt.savefig('demanda_negativa_dia.png', dpi=300)

    plt.show()
    
    return 
```

```python
day = 28
station = 49

demanda_negativa_plot(day, station)
```

```python
def demanda_t(day, station):
    
    grouped_days = data.groupby(data['Inicio_del_viaje'].dt.day)

    day_n = grouped_days.get_group(day)
    
    ## Negative Demand
    day_n_station_n = day_n.loc[day_n['Origen_Id'] == station]
    
    hourly_counts = day_n_station_n.groupby(day_n_station_n['Inicio_del_viaje'].dt.hour)['Inicio_del_viaje'].count()
    
    grouped_days = data.groupby(data['Inicio_del_viaje'].dt.day)

    try:
        midnight = pd.Series([hourly_counts[0]], index = [24])
                
    except:
        
        midnight = pd.Series([0], index = [24])
        
    hours_range = pd.Series(range(5, 25))

    hourly_counts = hourly_counts.reindex(hours_range, fill_value = 0)
    
    hourly_counts_negative = pd.concat([hourly_counts,midnight])

    ## Positive Demand

    day_n_station_n = day_n.loc[day_n['Destino_Id'] == station]
    
    hourly_counts = day_n_station_n.groupby(day_n_station_n['Inicio_del_viaje'].dt.hour)['Inicio_del_viaje'].count()
    
    try:
        midnight = pd.Series([hourly_counts[0]], index = [24])
                
    except:
        
        midnight = pd.Series([0], index = [24])
        
    hours_range = pd.Series(range(5, 25))

    hourly_counts = hourly_counts.reindex(hours_range, fill_value = 0)
    
    hourly_counts_positive = pd.concat([hourly_counts,midnight])
    
    
    #return hourly_counts_negative[7:15]
    return list(hourly_counts_positive[7:15] - hourly_counts_negative[7:15])
```

## With periods

```python
def total_demand(day, station, period):
    
    grouped_days = data.groupby(data['Inicio_del_viaje'].dt.day)

    day_n = grouped_days.get_group(day)
    
    ## Negative Demand
    day_n_station_n = day_n.loc[day_n['Origen_Id'] == station]
    
    hourly_counts = day_n_station_n.groupby(day_n_station_n['Inicio_del_viaje'].dt.hour)['Inicio_del_viaje'].count()
    
    try:
        midnight = pd.Series([hourly_counts[0]], index = [24])
                
    except:
        
        midnight = pd.Series([0], index = [24])
        
    hours_range = pd.Series(range(5, 25))

    hourly_counts = hourly_counts.reindex(hours_range, fill_value = 0)
    
    hourly_counts_negative = pd.concat([hourly_counts,midnight])

    ## Positive Demand

    day_n_station_n = day_n.loc[day_n['Destino_Id'] == station]
    
    hourly_counts = day_n_station_n.groupby(day_n_station_n['Inicio_del_viaje'].dt.hour)['Inicio_del_viaje'].count()
    
    try:
        midnight = pd.Series([hourly_counts[0]], index = [24])
                
    except:
        
        midnight = pd.Series([0], index = [24])
        
    hours_range = pd.Series(range(5, 25))

    hourly_counts = hourly_counts.reindex(hours_range, fill_value = 0)
    
    hourly_counts_positive = pd.concat([hourly_counts,midnight])
    
    if period == 1:
        return list(hourly_counts_positive[:7] - hourly_counts_negative[:7])
    elif period == 2:
        return list(hourly_counts_positive[7:15] - hourly_counts_negative[7:15])
    elif period == 3:
        return list(hourly_counts_positive[15:] - hourly_counts_negative[15:])
    else:
        return None
```
